# 1 Total Suicide In Various Countries Between This Time Period.
       Total Suicide According To Differnet Countries

```python
df1=df.groupby(['country'])['country','suicides_no'].sum()
df1
```

```python
df1['suicides_no'].plot()
```

# 2 Number Of Suicides In Accordance To Sex.
      Distribution Of Suicides For Different Sex.

```python
df.groupby(['sex'])['sex','suicides_no'].sum()
```

```python
sns.countplot(x='sex',data=df)
```

# 3 Suicide Rate And Number Of Suicides According To The Age Group(for Kids,for Adults).
      Suicide Details According To Different Age Groups.

```python
df3=df.groupby(['age'])['age','suicides_no','population'].sum()
df3['suicide_rate']=df3['suicides_no']*100000/df3['population']
df3
```

```python
df3['suicide_rate'].plot()
```

# 4 Suicide Rate In Relation With Generation All Over The Globe.
      Displaying The Total Number of Suicide Rate According To Different Generations.

```python
df4=df.groupby(['generation'])['generation','suicides_no','population'].sum()
df4['suicide_rate']=df4['suicides_no']*100000/df4['population']
df4
```

```python
df4['suicide_rate'].plot()
```

# 5 Sorting The Data On Basis Of  Suicide Rate In Accoradance To Year.
      Sorting Some Of My Data as To Get the flow of decrease in suicide rate according to year.

```python
df5=df.groupby(['year'])['year','suicides_no','population'].sum()
df5['suicide_rate']=df5['suicides_no']*100000/df5['population']
df5.sort_values(by='suicide_rate',ascending=False)
```

```python
df5['suicide_rate'].plot()
```

# 6 Calulating Mean Of The Number Of Suicides Per Year.
      Calculation the mean of the total suicides per year.

```python
df.groupby(['year'])['year','suicides_no'].mean()
#displaying the mean of number of suicides per year
```

# 9 To Find The Total GDP Per Capital.
      Hearby Displying My GDP per capita as it displays the Nation's living Standard and also is Flexible With The Change In Population.

# 10 Representing Suicides Per 100K Population For Each Year.
       Hearby Representing Graphically Suicides per 100k of population per year.

```python
df7=df.groupby('year')['year','suicides/100k pop'].sum()
df7['suicides/100k pop'].plot()
```

# 11 Displaying The Total HDI For Year.
       Displaying My Human Developmnet Index for each year.
